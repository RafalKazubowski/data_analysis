##### convert to bool

how to identify columns with bool values?

```python
# max 2 unique values in the column: true / false ; 0 / 1 ; yes / no etc
# or 1 unique value and NaN


col_bool = []

for col in mazda.columns:
    if mazda[col].nunique() == 1 or mazda[col].nunique() == 2: 
        for val in mazda[col].unique():
            if val in ['yes', 'no', True, False, 1, 0, 'Tak', 'Nie']:
                col_bool.append(col)
col_bool 
```

```python
# create function
def conv_to_bool (df):
    col_bool = []
    
    for col in df.columns:
        if df[col].nunique() == 1 or df[col].nunique() == 2: 
            for val in df[col].unique():
                if val in ['yes', 'no', True, False, 1, 0,'Tak', 'Nie']:
                    col_bool.append(col)
                    
    for col in col_bool:
        df[col] = df[col].astype('bool')
    return        
```

```python
conv_to_bool(mazda)
```

##### convert to datetime

```python
mazda[['data-dodania']].head(5)
```

```python
mazda['data-dodania'].replace({'czerwca': 'June'},regex=True, inplace=True)
mazda['data-dodania'].replace({'maja': 'May'},regex=True, inplace=True)
```

```python
mazda['data-dodania'] = mazda['data-dodania'].apply(lambda x: datetime.strptime(x, '%H:%M, %d %B %Y'))
```
